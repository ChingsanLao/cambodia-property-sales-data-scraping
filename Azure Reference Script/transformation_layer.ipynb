### TRANSFORMATIONS

##### Clean Amenities Dimension

```python
df_amenities.display()
```

```python
df_amenities_cleaned = df_amenities.select(
    col("amenity_id").alias("AmenityID"),
    col("amenity_name").alias("AmenityName"),
    col("property_id").alias("PropertyID")
)
```

```python
df_amenities_cleaned.write.format("parquet").mode("overwrite").save("abfss://silver@dwprojectstorage.dfs.core.windows.net/Dim_Amenities")
```

```python
df_amenities_cleaned.display()
```

##### Clean Location Dimension

```python
df_location.display()
```

```python
df_location_cleaned = df_location.select(
    col("location_id").alias("LocationID"),
    col("address").alias("Address"),
    col("commune").alias("Commune"),
    col("district").alias("District"),
    col("city").alias("City")
)
```

```python
df_location_cleaned.write.format("parquet").mode("overwrite").save("abfss://silver@dwprojectstorage.dfs.core.windows.net/Dim_Location")
```

```python
df_location_cleaned.display()
```

##### Clean Property Dimension

```python
df_property.display()
```

```python
df_property_cleaned = df_property.select(
    col("property_id").alias("PropertyID"),
    col("property_name").alias("PropertyName"),
    col("property_types").alias("PropertyType"),
    col("floor_level").alias("FloorLevel"),
    col("completion_year").alias("CompletionYear"),
    col("facing").alias("Facing")
)
```

```python
from pyspark.sql.functions import when, col

# Add a new column 'availability' based on the values in 'property_name'
df_property_cleaned = df_property.withColumn(
    "availability",
    when(col("property_name").contains("for Sale & Rent"), "Both Sale & Rent")
    .when(col("property_name").contains("for Sale"), "For Sale")
    .when(col("property_name").contains("for Rent"), "For Rent")
    .otherwise("Unknown")  # Handle cases where the availability type is not specified
)
```

```python
df_property_cleaned.display()
```

```python
df_property_cleaned.write.format("parquet").mode("overwrite").save("abfss://silver@dwprojectstorage.dfs.core.windows.net/Dim_Property")
```

##### Clean Time Dimension

```python
df_time.display()
```

```python
df_time_cleaned = df_time.select(
    col("date_id").alias("DateID"),
    col("date").alias("Date"),
    col("day_of_week").alias("DayOfWeek"),
    col("month").alias("Month"),
    col("quarter").alias("Quarter"),
    col("year").alias("Year")
)
```

```python
df_time_cleaned.write.format("parquet").mode("overwrite").save("abfss://silver@dwprojectstorage.dfs.core.windows.net/Dim_Time")
```

```python
df_time_cleaned.display()
```

##### Clean Property Listings Fact Table

```python
df_proplisting.display()
```

```python
fact_propertylist_cleaned = df_proplisting.select(
    col("property_id").alias("PropertyID"),
    col("location_id").alias("LocationID"),
    col("listed_date").alias("ListedDate"),
    col("updated_date").alias("UpdatedDate"),
    col("price").alias("Price"),
    col("floor_area").alias("FloorArea"),
    col("land_area").alias("LandArea"),
    col("bedrooms").alias("Bedrooms"),
    col("bathrooms").alias("Bathrooms"),
)
```
